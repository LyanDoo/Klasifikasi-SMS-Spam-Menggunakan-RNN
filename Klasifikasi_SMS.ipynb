```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset_sms_spam_v1.csv')
df.head()
```

```python
df.info()
```

```python
df['Teks'] = df['Teks'].str.lower()
df.head()
```

```python
df['Teks'] = df['Teks'].str.replace('[^a-zA-Z ]','')
df['Teks']
```

```python
df['Teks'] = df['Teks'].str.strip()
df['Teks']
```

```python
df.head()
```

```python
df['Teks'][4]
```

```python
teks_corpus = list(df['Teks'])
teks_corpus
```

```python
len(teks_corpus)
```

```python
teks_label = df['label'].values
teks_label
```

```python
teks_label.shape
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer(num_words=500,oov_token='-')
tokenizer.fit_on_texts(teks_corpus)
tokenizer.word_index
```

```python
sequences = tokenizer.texts_to_sequences(teks_corpus)
sequences
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

sequences = pad_sequences(sequences,padding='post',truncating='post')
sequences
```

```python
import re

def text_transform_pipe(text,tokenizer):
  text = text.lower()
  text = re.sub(r'[^a-zA-Z ]', '', text)
  text = text.strip()
  text = [text]
  seq = tokenizer.texts_to_sequences(text)
  seq = pad_sequences(seq,padding='post',truncating='post',maxlen=59)
  result = tf.one_hot(indices=seq,depth=499)
  return result
```

```python
sequences.shape
```

```python
sequences[0]
```

```python
sequences = tf.one_hot(indices=sequences,depth=sequences.max(),axis=-1)
sequences.shape
```

```python
sequences[0]
```

```python
teks_label_oh = tf.one_hot(indices=teks_label,depth=teks_label.max()+1)
teks_label_oh.shape
```

```python
teks_label_oh
```

```python
X_train = sequences[:900]
X_test = sequences[900:]
y_train = teks_label_oh[:900]
y_test = teks_label_oh[900:]
print('X Train shape =',X_train.shape)
print('X Test shape =',X_test.shape)
print('y Train shape =',y_train.shape)
print('y test shape =',y_test.shape)
```

```python
def rnn_model():
  input = tf.keras.layers.Input(shape=(59,499))
  x = tf.keras.layers.SimpleRNN(100,dropout=0.5)(input)
  output = tf.keras.layers.Dense(3,activation='softmax')(x)
  return tf.keras.Model(inputs=input,outputs=output)
```

```python
rnn_mod = rnn_model()
```

```python
rnn_mod.summary()
```

```python
rnn_mod.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0003),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
rnn_mod.fit(X_train,y_train,validation_data=[X_test,y_test],epochs=50)
```

```python
def prediksi(text,model,tokenizer):
  x = text_transform_pipe(text,tokenizer)
  hasil = model.predict(x)
  ind = hasil.argmax()
  if (ind == 0):
    print('"'+text+'"'+' << SMS normal dengan keyakinan '+str(hasil[0,ind]*100)+'%')
  elif (ind == 1):
    print('"'+text+'"'+' << SMS penipuan dengan keyakinan '+str(hasil[0,ind]*100)+'%')
  elif (ind == 2):
    print('"'+text+'"'+' << SMS Promo dengan keyakinan '+str(hasil[0,ind]*100)+'%')
```

```python
prediksi("Promo ini hanya anda bisa temukan di telkomsel",rnn_mod,tokenizer)
prediksi("SELAMAT ANDA BERUNTUNG MEMENANGKAN RP17.000.000",rnn_mod,tokenizer)
prediksi("dek kamu di rumah? mama minta pulsa dong, mama di kantor polisi soalnya",rnn_mod,tokenizer)
prediksi("Sayang kamu lagi ngapain sih, kok chatku gk dibales :(",rnn_mod,tokenizer)
```
